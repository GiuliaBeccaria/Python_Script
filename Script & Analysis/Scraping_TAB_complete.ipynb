```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time

driver = webdriver.Chrome()
driver.maximize_window()
url = 'https://www.nba.com/stats'
driver.get(url)
time.sleep(1)
teams_button = driver.find_element(By.XPATH, "//button[@class='SubNavItem_subnavLink__pF1m7']")
teams_button.click()
traditional_stats = driver.find_element(By.XPATH, "//a[@data-text='Traditional Stats']")
traditional_stats.click()
time.sleep(5)
element = driver.find_element(By.XPATH, "//button[@class='onetrust-close-btn-handler banner-close-button ot-close-link' and text()='Continue without Deciding']")
element.click()
time.sleep(5)
element = driver.find_element(By.XPATH, "//span[@class='StatsQuickNavSelector_selector__LkTyQ' and text()='Players']")
element.click()
time.sleep(5)
element = driver.find_element(By.XPATH, "//a[@class='StatsQuickNavSelector_link__yl1f2' and @data-active='false' and @role='button' and @tabindex='0' and @href='/stats/teams/traditional']")
element.click()
time.sleep(5)

advanced_filters_button = driver.find_element(By.XPATH, "//button[@class='ArrowToggleButton_arrowButton__9RpIv StatsAdvancedFiltersPanel_safArrow__EqRgu']")
advanced_filters_button.click()

position_filter = driver.find_element(By.XPATH, "//label[p[text()='Position']]//select")
position_select = Select(position_filter)

all_data = []

year_dropdown = driver.find_element(By.XPATH, "//select[@class='DropDown_select__4pIg9']")
year_select = Select(year_dropdown)
anni_disponibili = [option.get_attribute("value") for option in year_select.options]

for anno in anni_disponibili:
    year_select.select_by_value(anno)
    time.sleep(5)

    position_data = {"All Positions": []}

    for option in position_select.options: 
        position = option.get_attribute("value")
        position_select.select_by_value(position)
        get_stats_button = driver.find_element(By.XPATH, "//button[@data-btn-variant='primary' and text()='Get Stats']")
        get_stats_button.click()
        time.sleep(5)
        
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')
        table = soup.find('table', class_='Crom_table__p1iZz')
        data = []
        rows = table.find_all('tr')[1:] 
        for row in rows:
            cols = row.find_all('td')
            cols = [col.get_text() for col in cols]
            data.append([anno] + [position] + cols)

        position_data[position] = data  

    all_data.extend(position_data["All Positions"])

    for position, data in position_data.items():
        if position != "All Positions":
            all_data.extend(data)

driver.quit()

column_names = ["Anno", "Position", "id", "TEAM", "GP", "W", "L", "WIN%", "MIN", "PTS", "FGM", "FGA", "FG%", "3PM", "3PA", "3P%", "FTM", "FTA", "FT%",
                "OREB", "DREB", "REB", "AST", "TOV", "STL", "BLK", "BLKA", "PF", "PFD", "+/-"]  # Sostituisci con i nomi delle tue colonne

df = pd.DataFrame(all_data, columns=column_names)
print(df)
```

## Scraping with comments

```python
# Import necessary libraries
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time

# Initialize a Selenium Chrome web driver
driver = webdriver.Chrome()
driver.maximize_window()

# Define the URL to visit
url = 'https://www.nba.com/stats'
driver.get(url)

# Pause for 1 second to allow the page to load
time.sleep(1)

# Locate and click the "Teams" button on the webpage
teams_button = driver.find_element(By.XPATH, "//button[@class='SubNavItem_subnavLink__pF1m7']")
teams_button.click()

# Locate and click the "Traditional Stats" option
traditional_stats = driver.find_element(By.XPATH, "//a[@data-text='Traditional Stats']")
traditional_stats.click()

# Pause for 5 seconds
time.sleep(5)

# Handle a pop-up element with the text "Continue without Deciding"
element = driver.find_element(By.XPATH, "//button[@class='onetrust-close-btn-handler banner-close-button ot-close-link' and text()='Continue without Deciding']")
element.click()

# Pause for 5 seconds
time.sleep(5)

# Locate and click the "Players" option in the quick navigation menu
element = driver.find_element(By.XPATH, "//span[@class='StatsQuickNavSelector_selector__LkTyQ' and text()='Players']")
element.click()

# Pause for 5 seconds
time.sleep(5)

# Locate and click the "Teams" option in the quick navigation menu
element = driver.find_element(By.XPATH, "//a[@class='StatsQuickNavSelector_link__yl1f2' and @data-active='false' and @role='button' and @tabindex='0' and @href='/stats/teams/traditional']")
element.click()

# Pause for 5 seconds
time.sleep(5)

# Locate and click the "Advanced Filters" button
advanced_filters_button = driver.find_element(By.XPATH, "//button[@class='ArrowToggleButton_arrowButton__9RpIv StatsAdvancedFiltersPanel_safArrow__EqRgu']")
advanced_filters_button.click()

# Locate the filter for "Position" and create a Select object to interact with it
position_filter = driver.find_element(By.XPATH, "//label[p[text()='Position']]//select")
position_select = Select(position_filter)

# Initialize a list to store data
all_data = []

# Locate the year dropdown and create a Select object
year_dropdown = driver.find_element(By.XPATH, "//select[@class='DropDown_select__4pIg9']")
year_select = Select(year_dropdown)

# Get the available years as a list of values
anni_disponibili = [option.get_attribute("value") for option in year_select.options]

# Loop through each available year
for anno in anni_disponibili:
    year_select.select_by_value(anno)
    time.sleep(5)

    # Initialize a dictionary to store position data for the current year
    position_data = {"All Positions": []}

    # Loop through each position in the filter
    for option in position_select.options:  # Including "All Positions"
        position = option.get_attribute("value")

        # Select the position
        position_select.select_by_value(position)

        # Click the "Get Stats" button to update statistics
        get_stats_button = driver.find_element(By.XPATH, "//button[@data-btn-variant='primary' and text()='Get Stats']")
        get_stats_button.click()

        # Wait for the data to update (you can add a custom wait time)
        time.sleep(5)

        # Extract the page source
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, 'html.parser')

        # Find the data table (make sure the XPATH is correct)
        table = soup.find('table', class_='Crom_table__p1iZz')

        # Extract data from the table (ensure the table structure matches the code)
        data = []
        rows = table.find_all('tr')[1:]  # Skip the header
        for row in rows:
            cols = row.find_all('td')
            cols = [col.get_text() for col in cols]
            data.append([anno] + [position] + cols)

        position_data[position] = data  # Add data to the position in the dictionary

    # Add data for the current year to the main list
    all_data.extend(position_data["All Positions"])

    for position, data in position_data.items():
        if position != "All Positions":
            all_data.extend(data)

# Close the Selenium driver at the end
driver.quit()

# Define column names for the DataFrame
column_names = ["Year", "Position", "ID", "TEAM", "GP", "W", "L", "WIN%", "MIN", "PTS", "FGM", "FGA", "FG%", "3PM", "3PA", "3P%", "FTM", "FTA", "FT%",
                "OREB", "DREB", "REB", "AST", "TOV", "STL", "BLK", "BLKA", "PF", "PFD", "+/-"]  # Replace with your column names

# Create the final DataFrame
df = pd.DataFrame(all_data, columns=column_names)

# Print the final DataFrame
print(df)
```
